### Exploring the data

```python
import pandas as pd

top20_deathtoll = pd.read_csv('top20_deathtoll.csv')
print(top20_deathtoll)
```

### Design for the audience

```python
import matplotlib.pyplot as plt
%matplotlib inline  

fig, ax = plt.subplots(figsize=(4.5, 6)) # mobile-friendly 
ax.barh(top20_deathtoll["Country_Other"], top20_deathtoll["Total_Deaths"])
plt.show()
```

Now, we will try to redesign the graph by the following steps:
- Maximize the data ink
- Add details such as titles, labels, etc.

Note that we must keep the core data-inks that includes the information:
- The bars
- The x labels (the number of total deaths)
- The y labels (the country names)



```python
# Maximize the data ink by erasing non-data ink -- which are the ticks and axes
fig, ax = plt.subplots(figsize=(4.5, 6)) 
ax.barh(top20_deathtoll["Country_Other"], top20_deathtoll["Total_Deaths"])

for location in ['left', 'right', 'bottom', 'top']: # remove the spines
   ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False) # remove the ticks



plt.show()
```

```python
# Maximize the data ink by erasing redundant data-ink 
fig, ax = plt.subplots(figsize=(4.5, 6)) 
ax.barh(top20_deathtoll["Country_Other"], top20_deathtoll["Total_Deaths"], 
        height = 0.45)

for location in ['left', 'right', 'bottom', 'top']: # remove the spines
   ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False) # remove the ticks

ax.set_xticks([0, 150000, 300000]) # erase redundant data-ink

plt.show()
```

Notice that the x-label seems a bit incoherent from the overall features of the graph.
Let's try to move the x-tick to the top and try to draw the audience's attention to the bar by coloring them in red

```python
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],top20_deathtoll['Total_Deaths'],
         height=0.45, color='#af0b1e') # in red

for location in ['left', 'right', 'top', 'bottom']:
     ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.set_xticks([0, 150000, 300000])
ax.xaxis.tick_top() # move the xtick to the top
ax.tick_params(top=False, left=False)

plt.show()
```

Now, let's add the title and some labels to give the readers some information that we want to report

```python
# Give the readers more information by the title and subtitle
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],top20_deathtoll['Total_Deaths'],
         height=0.45, color='#af0b1e') # in red

for location in ['left', 'right', 'top', 'bottom']:
     ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.set_xticks([0, 150000, 300000])
ax.xaxis.tick_top() # move the xtick to the top
ax.tick_params(top=False, left=False)

ax.text(x=-40000, y=24, s='The Death Toll Worldwide Is 1.5M+', size=15, weight='bold') # data source: Worldometers
ax.text(x=-30000, y=22.5, s='Top 20 countries by death toll (December 2020)', size=11)

plt.show()
```

Since one of our titles is 'The Death Toll Worldwide Is 1.5M+', let's create a vertical line at value x = 150000

```python
# Final touch
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(top20_deathtoll['Country_Other'],top20_deathtoll['Total_Deaths'],
         height=0.45, color='#af0b1e') # in red

for location in ['left', 'right', 'top', 'bottom']:
     ax.spines[location].set_visible(False)
ax.tick_params(bottom=False, left=False)
ax.set_xticks([0, 150000, 300000])
ax.xaxis.tick_top() # move the xtick to the top
ax.tick_params(top=False, left=False)

ax.text(x=-40000, y=24, s='The Death Toll Worldwide Is 1.5M+', size=15, weight='bold') # data source: Worldometers
ax.text(x=-30000, y=22.5, s='Top 20 countries by death toll (December 2020)', size=11)

ax.axvline(x=150000, ymin=0.045, c='grey', alpha=0.5)

plt.show()
```
